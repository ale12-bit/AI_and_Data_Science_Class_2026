Q-3: arithmetic-program

```python

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "undefined"
modulus = int(num1) % int(num2) if num2 != 0 else "undefined"

print(f"\nResults for numbers {num1:.2f} and {num2:.2f}:")
print(f"Addition:       {addition:.2f}")
print(f"Subtraction:    {subtraction:.2f}")
print(f"Multiplication: {multiplication:.2f}")
print(f"Division:       {division}")
print(f"Modulus:        {modulus}")
```
